```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("/content/co2_emissions.csv")

# Use correct column names
years = data["Year"]
emissions = data["CO2_Emissions"]

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(years, emissions, marker="o", linestyle="-", linewidth=2)

# Add labels and title
plt.xlabel("Year")
plt.ylabel("CO2 Emissions (Million Metric Tons)")
plt.title("Global CO2 Emissions Over Time")

# Add grid and scaling
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()

# Show chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("/content/student_scores.csv")

# Group by Exam_Number and take average score (in case multiple students)
avg_scores = data.groupby("Exam_Number")["Score"].mean()

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(avg_scores.index, avg_scores.values, marker="o", linestyle="-", linewidth=2, color="blue")

# Add labels and title
plt.xlabel("Exam Number")
plt.ylabel("Average Score")
plt.title("Student Performance Over Exams")

# Add grid
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()

# Show chart
plt.show()

# Insights
print("Insights:")
if avg_scores.iloc[-1] > avg_scores.iloc[0]:
    print(" Students show improvement over time.")
else:
    print(" No clear improvement trend observed.")

print(f" Initial Average Score: {avg_scores.iloc[0]:.2f}")
print(f" Final Average Score: {avg_scores.iloc[-1]:.2f}")
```
